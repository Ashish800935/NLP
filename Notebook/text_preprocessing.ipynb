```python
## Remove URL
def remove_url(text):
    pattern = re.compile(r'http\S+|www\S+')
    return pattern.sub(r'', text)
```

```python
text1 = "Visit https://www.google.com for searching information."

text2 = "Watch tutorials at https://www.youtube.com/watch?v=abc123"

text3 = "My website is www.ashishkumar.com and it contains projects."

text4 = "Check both https://github.com and www.kaggle.com for learning."

text5 = "This movie is amazing! More details at https://www.imdb.com/title/tt0111161/"
```

```python
print(remove_url(text1))
print(remove_url(text2))
print(remove_url(text3))
print(remove_url(text4))
print(remove_url(text5))
```

```python
import string,time
string.punctuation
```

```python
exclude = string.punctuation
```

```python
def remove_punctuation(text):
    for char in exclude:
        text = text.replace(char, '')
    return text
```

```python
text = "Hello!!! How are you?"
start_time = time.time()
print(remove_punctuation(text))
end_time = time.time()
print(f"Execution time: {end_time - start_time} seconds")
```

```python
def remove_punctuation1(text):
    return text.translate(str.maketrans('', '', exclude))
```

```python
text = "Hello!!! How are you?"
start_time = time.time()
print(remove_punctuation1(text))
end_time = time.time()
print(f"Execution time: {end_time - start_time} seconds")
```

```python
def chat_conversion(text):
    words = text.split()
    new_words = []
    for word in words:
        if word in chat_words:
            new_words.append(chat_words[word])
        else:
            new_words.append(word)
    return ' '.join(new_words)
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
print(stop_words)
```

```python
def remove_stopwords(text):
    new_text=[]
    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x=new_text[:]
    new_text.clear()
    return ' '.join(x)    
```

```python
text = "I am learning NLP from CampusX"

print(remove_stopwords(text))
```

```python
def remove_emojis(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

```python
remove_emojis("Hello 😊! How are you? 🚀")
```

```python
from nltk.stem import PorterStemmer
```

```python
ps=PorterStemmer()
def stem_word(word):
    return " ".join([ps.stem(w) for w in word.split()])
```

```python
sample_text = "running runner runs"
print(stem_word(sample_text))
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
sentence = "The boys were playing football in the playground."
punctuations="?:!.,;"
sentence_words=word_tokenize(sentence)
for word in sentence_words:
    if word in punctuations:
        sentence_words.remove(word)

sentence_words
print("{0:20} {1:20} ".format("Word", "Lemma"))
for word in sentence_words:
    print("{0:20} {1:20}".format(word, wordnet_lemmatizer.lemmatize(word,pos='v')))        
```
