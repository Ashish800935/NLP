```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({'text':['people watch campusx','campusx watch campusx','people write comment','campusx write comment'],'output':[1,1,0,0]})
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
```

```python
bow=cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
print(bow[3].toarray())
```

```python
cv.transform(["campusx watch and write comment of campusx"]).toarray()
```

```python
## n-grams
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(ngram_range=(2,2)) # (1,1) for unigram, (2,2) for bigram, (3,3) for trigram, (1,2) for unigram and bigram
bow=cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
print(bow[3].toarray())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(ngram_range=(1,3)) #(1,3) for unigram, bigram and trigram
```

```python
bow=cv.fit_transform(df['text'])
print(len(cv.vocabulary_))
```

## tf-idf

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
tfidf.fit_transform(df['text']).toarray()
```

```python
print(tfidf.idf_)
print(tfidf.get_feature_names_out())
```
